```python
## 假如我們不想把資料載到自己的電腦裡?
# 把連結填入
target_url = 'https://raw.githubusercontent.com/vashineyu/slides_and_others/master/tutorial/examples/imagenet_urls_examples.txt'
```

```python
import re
import requests
response = requests.get(target_url)

#伺服器回應的狀態碼
print(response.status_code) #回應200代表正常

# 檢查狀態碼是否 OK
try:    
    if response.status_code == requests.codes.ok:
      print("OK")
except:
    print("伺服器無回應")
    
data = response.text #<class 'str'>
# 用 request 傳送回來的資料不會認得斷行符號
print(len(data))
data[0:100]
```

```python
# python內建的split不能一次分割多個，但是re.split卻可以。 https://blog.csdn.net/samsam2013/article/details/78456185
import re
# 找到換行符號，用該符號做字串分割後，把它拿掉
data = re.split('[\n\t]',data)
print(len(data))
data[0:10]
```

## 將 txt 轉成 pandas dataframe

```python
import pandas as pd
import requests
import re
txt_rul = 'https://raw.githubusercontent.com/vashineyu/slides_and_others/master/tutorial/examples/imagenet_urls_examples.txt'
text = requests.get(txt_rul).text

# 利用re正規表示式拿掉亂碼，只留下網址。
x = re.compile('[a-z0-9]+_[0-9]+') # 去掉n00433802_4955這類的亂碼
text = re.sub(x, '', text)
text = re.sub('\t', '',text)
arrange_data = text.split('\n')
#arrange_data = arrange_data.append(a)
#for a in text:
    #a = a.replace('\n','').split(',')
    #a = re.sub('\n', ',',text).split(',')
    
    #arrange_data.append(a)
#print(arrange_data)
df = pd.DataFrame(arrange_data)
df.head(6) # 列出前六個
```
